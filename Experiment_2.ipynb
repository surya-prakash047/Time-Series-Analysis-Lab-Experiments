```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the uploaded CSV file
file_name = 'data.csv'  # Ensure the file you upload is named 'data.csv'
data = pd.read_csv(file_name, parse_dates=['Date'], index_col='Date')
```

```python
# Display the column names to identify the actual name
print("Available columns:", data.columns)
```

```python
# Replace 'Value' with the actual name of the numerical column in your dataset
numerical_column = data.columns[0]  # Automatically selects the first column; adjust if needed
```

```python
# Ensure the data is sorted by date
data.sort_index(inplace=True)
```

```python
# Line Plot of Time Series Data
plt.figure(figsize=(12, 6))
plt.plot(data[numerical_column], label="Time Series Data")
plt.title("Line Plot of Time Series Data")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Visualizing Seasonality and Trends
monthly_data = data[numerical_column].resample("M").mean()  # Resampling to monthly frequency
plt.figure(figsize=(12, 6))
plt.plot(monthly_data, label="Monthly Average")
plt.title("Monthly Trends")
plt.xlabel("Date")
plt.ylabel("Average Value")
plt.legend()
plt.show()
```

```python
# Histogram of the values
plt.figure(figsize=(10, 5))
data[numerical_column].hist(bins=30, color='blue', alpha=0.7)
plt.title("Histogram of Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot of the values
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x=numerical_column, color='skyblue')
plt.title("Box Plot of Values")
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Simulating a time series dataset (e.g., daily temperature data over one year)
date_range = pd.date_range(start="2024-01-01", periods=365, freq="D")
temperature = 20 + 10 * np.sin(np.linspace(0, 2 * np.pi, 365)) + np.random.normal(0, 2, 365)

# Creating the DataFrame
data = pd.DataFrame({'Date': date_range, 'Temperature': temperature})

# Set Date as the index
data.set_index('Date', inplace=True)

# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Temperature'], label='Temperature', color='tab:blue')
plt.title("Daily Temperature Over One Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
